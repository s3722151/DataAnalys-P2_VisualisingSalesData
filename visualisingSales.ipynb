```python
#Help identify outliers
# Practical Data Science 
df['Sales'].dropna().plot(kind='box')
```

```python
# Bar plot - DF: Looking at Ship Mode
df['Segment'].value_counts().plot(kind='bar')
```

```python
# Bar plot - DF: Looking at Ship Mode
df['Ship Mode'].value_counts().plot(kind='bar')
```

```python
# Bar plot - DF: Looking at Ship Mode
df['Region'].value_counts().plot(kind='bar')
```

```python
# Create scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='Order Date',
    y='Ship Date',
    alpha=0.5,  # Makes points slightly transparent
    color='blue'
)

# Add a 45-degree reference line (where Order Date = Ship Date)
plt.plot(
    [df['Order Date'].min(), df['Order Date'].max()],
    [df['Order Date'].min(), df['Order Date'].max()],
    color='red',
    linestyle='--',
    label='Same-day shipping'
)

plt.title('Order Date vs. Ship Date (Scatter Plot)')
plt.xlabel('Order Date')
plt.ylabel('Ship Date')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 products by sales
top_products = df.groupby('Product Name')['Sales'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
top_products.sort_values().plot(kind='barh')
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 products by sales (excluding zero sales)
bottom_products = df.groupby('Product Name')['Sales'].sum().nsmallest(10)

plt.figure(figsize=(12, 6))
bottom_products.sort_values(ascending=False).plot(kind='barh')
plt.title('Bottom 10 Products by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
# Sales by category and sub-category
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Category', y='Sales', hue='Sub-Category', estimator=sum, ci=None)
plt.title('Sales by Product Category and Sub-Category')
plt.ylabel('Total Sales ($)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
